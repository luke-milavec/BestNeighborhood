### Education
In my personal notebook I am attempting to figure out how Education can help define the "best" neighborhood in Pittsburgh. This data set has a column of neighborhood names and other columns consist of pupulation and percentages of people with certain levels of education attainment. These attainments include: less than high school, high school graduate, assoc./prof. degree, bachelor's degree, and postgraduate degree. 


```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
education = pd.read_csv("./data/education-income.csv")
education
```

```python
education2 = education.drop(['Sector #', 'Population (2010)', 'Total Pop, 25 and older (2010)','Edu. Attainment: Less than High School (2010)', '1999 Median Income (\'99 Dollars)', '2009 Median Income (\'09 Dollars)', '1999 Median Income (\'11 Dollars)', '2009 Med. Income (\'13 Dollars)','Est. Pop. for which Poverty Calc. (2010)', 'Est. Pop. Under Poverty (2010)', 'Est. Percent Under Poverty (2010)'], axis=1)
education2.head(10)
```

```python
education2['Edu. Attainment: Postgraduate Degree (2010)'] = education2['Edu. Attainment: Postgraduate Degree (2010)'].str.rstrip('%').astype('float') / 100.0
education2   
```

```python
education2['Edu. Attainment: Bachelor\'s Degree (2010)'] = education2['Edu. Attainment: Bachelor\'s Degree (2010)'].str.rstrip('%').astype('float') / 100.0
education2                            
```

```python
education2['Edu. Attainment: High School Graduate (2010)'] = education2['Edu. Attainment: High School Graduate (2010)'].str.rstrip('%').astype('float') / 100.0
education2    
```

```python
education2['Edu. Attainment: Assoc./Prof. Degree (2010)'] = education2['Edu. Attainment: Assoc./Prof. Degree (2010)'].str.rstrip('%').astype('float') / 100.0
education2 
```

```python
education2.plot(x='Neighborhood', kind='bar', rot=90, figsize = (30,20))
```

```python
condenseEducation1 = education2.loc[(education2['Edu. Attainment: High School Graduate (2010)'] > 0.05) & (education2['Edu. Attainment: Assoc./Prof. Degree (2010)'] > 0.05) & (education2['Edu. Attainment: Bachelor\'s Degree (2010)'] > 0.05) & (education2['Edu. Attainment: Postgraduate Degree (2010)'] >0.05)]
condenseEducation1.plot(x='Neighborhood', kind='bar', rot=90, figsize = (20,10))
```

**deciding to just focus on postgraduate degree.**

```python
education3 = education2.drop(['Edu. Attainment: High School Graduate (2010)', 'Edu. Attainment: Assoc./Prof. Degree (2010)', 'Edu. Attainment: Bachelor\'s Degree (2010)'], axis=1)
education3['Edu. Attainment: Postgraduate Degree (2010)'] = education2['Edu. Attainment: Postgraduate Degree (2010)'].str.rstrip('%').astype('float') / 100.0
education3.head(10)
```

```python
condenseEducation = education3.loc[education3['Edu. Attainment: Postgraduate Degree (2010)'] > 0.15]
condenseEducation.plot(x='Neighborhood', kind='bar', rot=90, figsize = (20,10))
```

```python
len(condenseEducation['Neighborhood'].unique())
```

```python
condenseEducation.sort_values("Edu. Attainment: Postgraduate Degree (2010)", ascending=False).head
```

**shows south shore as having the highest percentage attaining postgraduate degrees**
